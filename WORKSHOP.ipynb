```python
Q1=dt['cogs'].quantile(0.25)
Q3=dt['cogs'].quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
lower_bound,upper_bound
```

```python
outliers = [x for x in dt['cogs'] if x < lower_bound or x > upper_bound]
outliers
```

```python
print("Q1:",Q1)
print("Q3:",Q3)
print("IQR:",IQR)
print("Lower Bound:",lower_bound)
print("Upper Bound:",upper_bound)
print("Outliers:",outliers)
```

```python
dt=dt[((dt['cogs']>=lower_bound)&(dt['cogs']<=upper_bound))]
dt
```
